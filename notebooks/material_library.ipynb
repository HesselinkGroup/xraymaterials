# Material library

**For overview of `xraymaterials` see `quickstart.ipynb`.**

The materials library contains elements 1-99 and some useful compounds and mixtures.  They can be discovered with autocomplete in Jupyter (hit the tab key!) or with the `list()` function.

## Elements

Pure elements.  Their default densities are taken from the `xraymaterials.elements` subpackage; the density is probably for the most prevalent phase at room temperature, but you should double-check.

## Compounds

The compounds subpackage collects materials with known chemical formulas.  Their densities are taken from various sources (documented in `xraymaterials.library.compounds.py` where applicable).  Some compounds may have density of 1 g/cc, the default in case the density is unknown.

## Mixtures

Complicated mixtures of elements and compounds according to recipes found in various sources.  The proportions and total densities were found in various sources online, documented as applicable in `xraymaterials.library.mixtures.py`.

```python
import xraymaterials

print("== All materials:")
print(xraymaterials.library.list())

print("\n== Elements:")
print(xraymaterials.library.elements.list())

print("\n== Compounds:")
print(xraymaterials.library.compounds.list())

print("\n== Mixtures:")
print(xraymaterials.library.mixtures.list())
```

## Material imports

All elements, compounds and mixtures are variables exposed in `xraymaterials.library`.

```python
import xraymaterials.library as xrl

energy_keV = np.logspace(np.log10(3.0), np.log10(400.0), 200)

plt.figure()
plt.loglog(energy_keV, xrl.water.mu(energy_keV), label='Water')
plt.loglog(energy_keV, xrl.gold.mu(energy_keV), label='Gold')
plt.loglog(energy_keV, xrl.fructose.mu(energy_keV), label='Fructose')
plt.loglog(energy_keV, xrl.sugar_maple.mu(energy_keV), label='Sugar Maple')
plt.loglog(energy_keV, xrl.cotton_clothes_packed.mu(energy_keV), label='Cotton clothes (packed)')
plt.grid()
plt.legend(loc='best')
plt.title("Absorption coefficients of several materials")


plt.figure()
plt.loglog(energy_keV, xrl.water.delta(energy_keV), label='Water')
plt.loglog(energy_keV, xrl.honey.delta(energy_keV), label='Honey')
plt.loglog(energy_keV, xrl.vinegar.delta(energy_keV), label='Vinegar')
plt.grid()
plt.legend(loc='best')
plt.title("Refractive index delta of several liquids")
```
